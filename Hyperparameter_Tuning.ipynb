```python
#grid search
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
X=iris.data
y=iris.target
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import numpy as np
```

```python
c_space=np.logspace(-5,8,15)
```

```python
c_space
```

```python
param_grid={'C':c_space}
```

```python
#model
lr=LogisticRegression(max_iter=1000)
```

```python
lrgrid=GridSearchCV(lr,param_grid,cv=5)
```

```python
lrgrid.fit(X,y)
```

```python
print('best params:' , lrgrid.best_params_)
print('best score:' , lrgrid.best_score_)
```

```python
from scipy.stats import randint
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_dist={
    'max_depth':[3,None],
    'max_features':randint(1,4),
    'criterion':['gini','entropy']
}
```

```python
dt=DecisionTreeClassifier()
```

```python
dt_random=RandomizedSearchCV(dt,param_dist,cv=5)
```

```python
dt_random.fit(X,y)
```

```python
print('best params:' , dt_random.best_params_)
print('best score:' , dt_random.best_score_)
```
